## 5. 거스름돈 계산기


역할: 너는 파이썬 튜터야

목표: 아래 문제를 파이썬으로 풀어줘.

문제: 거스름돈 계산기를 만들어줘. 사용자가 1만원 단위의 금액을 입력하면, 상품값을 제외한 거스름돈은 5천원, 1천원, 500원, 100원의 거스름돈을 받을 수 있도록 해줘.

입력: (예) 지불금액  : 20000 상품금액 : 15400

출력: (예)==거스름돈 ==
5,000원권 : 0장 1,000원권 : 4장 500원 동전 : 0개  100원 동전 : 3개 (각 권종당 줄바꿈으로 처리)

제약: 몫과 나머지 연산자 사용, 1000원 이상은 지폐 500원, 100원은 동전으로 처리, 사용자가 10,000원 단위로 입력할 수 있도록 제시.

요청: 코드에 대해 한 줄씩 설명하고 테스트 3개 실행


```python
# TODO: 거스름돈 계산기
```

```python
# 거스름돈 계산기 프로그램

# 1. 사용자로부터 지불 금액과 물품 금액을 입력받습니다.
# 지불 금액은 정수로 입력받아 payment 변수에 저장합니다.
payment = int(input("지불 금액을 만원 단위로 입력하세요 (예: 10000, 20000): "))
# 물품 금액은 정수로 입력받아 item_price 변수에 저장합니다.
item_price = int(input("물품 금액을 입력하세요 (예: 15400): "))

# 2. 지불 금액의 유효성을 검사합니다.
# 지불 금액이 10000의 배수가 아니면 오류 메시지를 출력하고 프로그램을 종료합니다.
# 만약 payment % 10000이 0이 아니면 (즉, 10000으로 나누어 떨어지지 않으면) 유효하지 않은 입력입니다.
if payment % 10000 != 0:
    print("오류: 지불 금액은 만원 단위로 입력해야 합니다.")
    # exit() 함수를 사용하여 프로그램 강제 종료 (Colab에서는 잘 동작하지 않을 수 있음)
    # 실제 환경에서는 다른 방식으로 오류를 처리하거나 재입력을 유도할 수 있습니다.
else:
    # 3. 거스름돈을 계산합니다.
    # 지불 금액에서 물품 금액을 빼서 거스름돈을 계산합니다.
    change = payment - item_price

    # 4. 거스름돈이 부족한 경우를 처리합니다.
    # 거스름돈이 0보다 작으면 (즉, 지불 금액이 물품 금액보다 적으면) 오류 메시지를 출력합니다.
    if change < 0:
        print("오류: 지불 금액이 물품 금액보다 부족합니다.")
    else:
        # 5. 각 권종별 거스름돈 매수를 계산하고 출력합니다.
        print("\n== 거스름돈 ==")

        # 5천원권 계산: 거스름돈을 5000으로 나눈 몫이 5천원권 매수입니다.
        # // 연산자는 정수 나눗셈을 수행합니다.
        num_5000 = change // 5000
        # 5천원권 매수를 계산한 후 남은 거스름돈을 다시 계산합니다.
        # % 연산자는 나눗셈의 나머지를 계산합니다.
        change = change % 5000
        print(f"5,000원권 : {num_5000}장")

        # 1천원권 계산: 남은 거스름돈을 1000으로 나눈 몫이 1천원권 매수입니다.
        num_1000 = change // 1000
        # 1천원권 매수를 계산한 후 남은 거스름돈을 다시 계산합니다.
        change = change % 1000
        print(f"1,000원권 : {num_1000}장")

        # 500원 동전 계산: 남은 거스름돈을 500으로 나눈 몫이 500원 동전 개수입니다.
        num_500 = change // 500
        # 500원 동전 개수를 계산한 후 남은 거스름돈을 다시 계산합니다.
        change = change % 500
        print(f"500원 동전 : {num_500}개")

        # 100원 동전 계산: 남은 거스름돈을 100으로 나눈 몫이 100원 동전 개수입니다.
        num_100 = change // 100
        # 100원 동전 개수를 계산한 후 남은 거스름돈을 다시 계산합니다.
        change = change % 100
        print(f"100원 동전 : {num_100}개")

        # 최종적으로 남은 거스름돈이 있다면 출력 (이론적으로는 100원 단위까지만 계산하므로 0이 되어야 함)
        if change > 0:
            print(f"주의: 나머지 잔돈 {change}원이 발생했습니다.")
```

## 8. 성적 분석 프로그램 만들기


### 문제 : 점수 5 개를 입력받아 리스트로 저장
출력: 평균/최대/최소/합계

도전: 90점 이상 개수 카운트하여 출력

"테스트 입력”을 함께 만들기


```python
# TODO: 성적 분석 프로그램 만들기
```

```python
a = [] # 점수들을 저장할 빈 리스트 초기화

# 사용자로부터 5개의 점수를 입력받아 리스트 'a'에 추가합니다. (for 문 사용)
for i in range(5):
  while True: # 올바른 점수를 입력받을 때까지 반복 (유효성 검사)
    try:
      score = int(input(f"{i+1}번째 점수를 입력하세요 (0~100): "))
      if 0 <= score <= 100:
        a.append(score)
        break # 유효한 점수이므로 while 루프 종료
      else:
        print("잘못된 입력값입니다. 점수는 0에서 100 사이여야 합니다. 다시 입력해주세요.")
    except ValueError:
      print("유효한 숫자를 입력해주세요.")

print(f"\n입력된 점수: {a}")

# 총점 계산
total_score = sum(a)
print(f"총점: {total_score}")

# 평균 계산 (실수형으로 출력)
# 리스트가 비어있지 않은 경우에만 평균을 계산합니다.
if len(a) > 0:
    average_score = total_score / len(a)
    print(f"평균: {average_score:.2f}") # 소수점 둘째 자리까지 출력
else:
    print("평균을 계산할 점수가 없습니다.")

# 최고점과 최저점 계산
if len(a) > 0:
    max_score = max(a)
    min_score = min(a)
    print(f"최고점: {max_score}")
    print(f"최저점: {min_score}")
else:
    print("최고점과 최저점을 계산할 점수가 없습니다.")

# 90점 이상인 점수의 개수 카운트 (for 문 사용)
count_above_90 = 0
for score_val in a: # 리스트의 각 요소를 직접 순회
  if score_val >= 90:
    count_above_90 += 1
print(f"90점 이상 점수 개수: {count_above_90}개")
```

### 문제 : 리스트에 할 일을 저장
add: 입력받아 append

list: 현재 목록 출력(번호 포함)

done: 번호를 입력받아 pop/remove로 삭제


```python
# TODO: TODO 리스트 만들기
```

```python
todo_list = [] # 할 일을 저장할 빈 리스트 초기화

while True:
    print("\n--- ToDo 리스트 관리 프로그램 ---")
    print("1. 할 일 추가")
    print("2. 할 일 삭제")
    print("3. 할 일 목록 보기")
    print("4. 종료")

    choice = input("원하는 작업을 선택하세요 (1-4): ")

    if choice == '1':
        task = input("추가할 할 일을 입력하세요: ")
        todo_list.append(task)
        print(f"'{task}'가 목록에 추가되었습니다.")
    elif choice == '2':
        if not todo_list:
            print("삭제할 할 일이 없습니다. 목록이 비어 있습니다.")
            continue # 리스트가 비어있으면 다시 메뉴로 돌아감

        print("\n--- 현재 할 일 목록 ---")
        for i, task in enumerate(todo_list):
            print(f"{i+1}. {task}")

        try:
            task_num_to_delete = int(input("삭제할 할 일의 번호를 입력하세요: "))
            if 1 <= task_num_to_delete <= len(todo_list):
                removed_task = todo_list.pop(task_num_to_delete - 1)
                print(f"'{removed_task}'가 목록에서 삭제되었습니다.")
            else:
                print("잘못된 번호입니다. 다시 시도해주세요.")
                continue # 잘못된 번호 입력 시 다시 메뉴로 돌아감
        except ValueError:
            print("유효한 숫자를 입력해주세요.")
            continue # 숫자가 아닌 값 입력 시 다시 메뉴로 돌아감
    elif choice == '3':
        if not todo_list:
            print("할 일 목록이 비어 있습니다.")
        else:
            print("\n--- 현재 할 일 목록 ---")
            for i, task in enumerate(todo_list):
                print(f"{i+1}. {task}")
    elif choice == '4':
        print("프로그램을 종료합니다.")
        break # 프로그램 종료
    else:
        print("잘못된 선택입니다. 1에서 4 사이의 숫자를 입력해주세요.")
        continue # 잘못된 선택 입력 시 다시 메뉴로 돌아감
```


## 10. 숫자 맞히기 게임(while)


### 요구사항
- 정답 숫자를 맞힐 때까지 반복 입력
- 힌트: 작으면 UP, 크면 DOWN
- 추가: 시도 횟수 카운트
- 무한 루프 방지: 시도 횟수 제한(max_tries) 같은 안전장치를 두어도 좋습니다.


```python
# TODO: 숫자 맞히기 게임
import random

target = random.randint(1, 100) # 1부터 100 사이의 무작위 정답 숫자 설정
max_tries = 10 # 최대 시도 횟수 설정
tries = 0 # 시도 횟수 초기화

print("1부터 100까지의 숫자 맞히기 게임! (기회: 10번)")

while True:
    try:
        guess = int(input("숫자를 입력하세요: "))
        tries += 1

        if guess == target:
            print(f"정답! {tries}번 만에 맞히셨습니다.")
            break # 정답을 맞히면 반복문 종료
        elif guess < target:
            print("UP")
        else:
            print("DOWN")

        if tries >= max_tries:
            print(f"아쉽지만 기회를 모두 소진했습니다. 정답은 {target}이었습니다.")
            break # 최대 시도 횟수를 넘기면 반복문 종료
    except ValueError:
        print("유효한 숫자를 입력해주세요.")
    except Exception as e:
        print(f"오류 발생: {e}")
        break
```

### 02. 선체 구조 부재 무게 등급 판별   [제어문]

문제: 선박에 사용되는 강판의 무게를 입력받아, 무게에 따라 크레인 운송 등급(A, B, C)을 판별하는 프로그램을 작성하세요.
조건: 50톤 이상은 'A등급', 20톤 이상 50톤 미만은 'B등급', 20톤 미만은 'C등급'으로 분류합니다.

```python
# 선체 구조 부재 무게 등급 판별 프로그램

while True:
    try:
        # 사용자로부터 강판의 무게를 입력받습니다.
        # float()을 사용하여 소수점 이하의 무게도 처리할 수 있도록 합니다.
        weight_input = input("강판의 무게를 입력하세요 (톤): ")
        weight = float(weight_input)

        # 음수 값에 대한 유효성 검사 추가
        if weight < 0:
            print("오류: 무게는 음수가 될 수 없습니다. 다시 입력해주세요.")
            continue # 음수면 다시 입력받도록 반복
        else:
            break # 유효한 양수 또는 0이 입력되면 반복 종료

    except ValueError:
        print("오류: 유효한 숫자를 입력해주세요.")
        continue # 숫자가 아닌 값 입력 시 다시 입력받도록 반복
    except Exception as e:
        print(f"예상치 못한 오류가 발생했습니다: {e}")
        break # 그 외 예외 발생 시 프로그램 종료

# 유효한 무게가 입력된 후에 등급을 판별합니다.
# 조건에 따라 등급을 판별합니다.
if weight >= 50: # 50톤 이상인 경우
    grade = 'A등급'
elif weight >= 20: # 20톤 이상 50톤 미만인 경우
    grade = 'B등급'
else: # 20톤 미만인 경우
    grade = 'C등급'

# 판별된 등급을 출력합니다.
print(f"강판의 무게: {weight}톤, 크레인 운송 등급: {grade}")
```

### 03.  강재 리스트 반복 출력 및 합계 [반복문]

문제: 리스트에 저장된 여러 개의 강재 무게 데이터를 반복문을 통해 하나씩 출력하고, 전체 강재의 총 무게를 계산하는 프로그램을 작성하세요.

```python
# 강재 무게 데이터 리스트 (사용자 입력)
steel_weights = [] # 강재 무게를 저장할 빈 리스트

while True:
    try:
        num_weights = int(input("입력할 강재의 개수를 입력하세요: "))
        if num_weights <= 0:
            print("오류: 강재의 개수는 1 이상이어야 합니다. 다시 입력해주세요.")
            continue
        break
    except ValueError:
        print("오류: 유효한 숫자를 입력해주세요.")

print("--- 각 강재의 무게 입력 ---")
for i in range(num_weights):
    while True:
        try:
            weight_input = input(f"{i+1}번째 강재의 무게를 입력하세요 (톤): ")
            weight = float(weight_input)

            if weight < 0:
                print("오류: 무게는 음수가 될 수 없습니다. 다시 입력해주세요.")
                continue # 음수면 다시 입력받도록 반복
            else:
                steel_weights.append(weight)
                break # 유효한 양수 또는 0이 입력되면 반복 종료
        except ValueError:
            print("오류: 유효한 숫자를 입력해주세요.")
            continue # 숫자가 아닌 값 입력 시 다시 입력받도록 반복

total_weight = 0 # 총 무게를 저장할 변수 초기화

print("\n--- 입력된 각 강재의 무게 ---")
# 리스트를 반복하면서 각 강재의 무게를 출력하고 총 무게에 더합니다.
for weight in steel_weights:
    print(f"강재 무게: {weight} 톤")
    total_weight += weight # 현재 강재 무게를 총 무게에 누적

print("--------------------------")
print(f"전체 강재의 총 무게: {total_weight:.1f} 톤")
```
