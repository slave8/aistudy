## 누적합/누적곱 함수로 만들어보기

````python
total = 0
fac = 1
for i in range(1, 6):
    total = total + i
    fac = fac * i
print("sum(1..5):", total)
print("fac(1..5):", fac)
````

````python
def sum(start, end) :
  total=0
  for i in range(start, end+1):
    total = total + i
  return total
def fac(start, end) :
  fac = 1
  for i in range(start, end+1):
    fac = fac * i
  return fac

num1=int(input("정수1 입력:"))
num2=int(input("정수2 입력:"))
print(f"sum({num1}..{num2}):", sum(num1,num2))
print(f"fac({num1}..{num2}):", fac(num1,num2))
````

````python
%%writefile "/content/drive/MyDrive/Colab Notebooks/contact.py"
contacts = {} # 연락처를 저장할 딕셔너리

def add(name, phone_number):
    # 연락처 추가 또는 수정 함수
    # 유효성 검사 로직은 메인 루프에서 처리되므로, 이 함수는 실제로 추가/수정만 수행합니다.
    if name in contacts:
        print(f"'{name}'의 연락처가 {contacts[name]}에서 {phone_number}로 수정되었습니다.")
    else:
        print(f"'{name}'의 연락처가 추가되었습니다.")
    contacts[name] = phone_number

def find(name):
    # 연락처 검색 함수
    phone_number = contacts.get(name) # get() 메소드를 사용하여 키가 없을 경우 None 반환
    if phone_number:
        print(f"'{name}'의 연락처: {phone_number}")
    else:
        print(f"'{name}'은(는) 연락처 목록에 없습니다.")

def list_contacts():
    # 전체 연락처 목록 출력 함수
    if not contacts:
        print("연락처 목록이 비어 있습니다.")
        return
    print("\n--- 연락처 목록 ---\n")
    for name, phone_number in contacts.items():
        print(f"이름: {name}, 번호: {phone_number}")
    print("\n-------------------\n")

def delete_contact(name):
    # 연락처 삭제 함수
    if name in contacts:
        del contacts[name]
        print(f"'{name}'의 연락처가 삭제되었습니다.")
    else:
        print(f"'{name}'은(는) 연락처 목록에 없습니다.")

# --- 메인 프로그램 시작 ---
while True:
    print("\n--- 연락처 관리 프로그램 ---")
    print("1. 연락처 추가/수정")
    print("2. 연락처 찾기")
    print("3. 전체 연락처 목록 보기")
    print("4. 연락처 삭제")
    print("5. 종료")

    choice = input("원하는 작업을 선택하세요 (1-5): ")

    if choice == '1': # 연락처 추가/수정
        name_to_add = ""
        phone_number_to_add = ""

        # 이름 입력 및 유효성 검사 루프
        while True:
            input_name = input("이름을 입력하세요: ")
            if input_name in contacts:
                print(f"경고: '{input_name}'은(는) 이미 존재하는 이름입니다. 현재 번호는 {contacts[input_name]}입니다.")
                action = input("이 연락처를 수정하시겠습니까? (y/n, 메뉴로 돌아가려면 'm'): ").lower()
                if action == 'y':
                    name_to_add = input_name
                    break # 이름 수정 확인, 번호 입력으로 진행
                elif action == 'm':
                    break # 메뉴로 돌아가기
                else:
                    print("다른 이름을 입력해주세요.")
            else:
                name_to_add = input_name
                break # 새 이름 입력 확인, 번호 입력으로 진행

        if not name_to_add: # If user chose to go back to menu from name input
            continue # Skip rest of this iteration and show main menu again

        # 전화번호 입력 및 유효성 검사 루프
        while True:
            input_phone_number = input("전화번호를 입력하세요: ")

            is_duplicate_number_for_other_name = False
            duplicate_name = ""
            for existing_name, existing_number in contacts.items():
                # 같은 번호이면서, 이름이 다른 경우 (현재 추가하려는 이름과 다른 경우)
                if existing_number == input_phone_number and existing_name != name_to_add:
                    is_duplicate_number_for_other_name = True
                    duplicate_name = existing_name # 중복된 번호를 가진 다른 사람의 이름 저장
                    break

            if is_duplicate_number_for_other_name:
                print(f"경고: '{input_phone_number}'은(는) 이미 다른 연락처 ('{duplicate_name}')에 등록된 번호입니다.")
                action = input(f"그래도 '{name_to_add}'의 번호로 이 번호를 저장하시겠습니까? (y/n, 메뉴로 돌아가려면 'm'): ").lower()
                if action == 'y':
                    phone_number_to_add = input_phone_number
                    break # 사용자가 중복 번호 사용을 허용, add 함수 호출로 진행
                elif action == 'm':
                    break # 메뉴로 돌아가기
                else:
                    print("다른 전화번호를 입력해주세요.")
            else:
                phone_number_to_add = input_phone_number
                break # 번호가 유니크하거나 같은 이름의 번호, add 함수 호출로 진행

        if not phone_number_to_add: # If user chose to go back to menu from phone number input
            continue # Skip rest of this iteration and show main menu again

        add(name_to_add, phone_number_to_add)

    elif choice == '2': # 연락처 찾기
        name = input("찾을 이름을 입력하세요: ")
        find(name)
    elif choice == '3': # 전체 연락처 목록 보기
        list_contacts()
    elif choice == '4': # 연락처 삭제
        name = input("삭제할 이름을 입력하세요: ")
        delete_contact(name)
    elif choice == '5': # 종료
        print("프로그램을 종료합니다.")
        break  # 프로그램 종료
    else:
        print("잘못된 선택입니다. 1에서 5 사이의 숫자를 입력해주세요.")
````

### 3-2. 각 학생의 평균 구하기


````python
def to_int(x):
    return int(x)

for row in rows:
    name = row["name"]
    korean = to_int(row["korean"])
    math = to_int(row["math"])
    english = to_int(row["english"])
    avg = (korean + math + english) / 3
    print(f"{name}: avg={avg:.2f}")
````

### 05. 선박 기자재 점검 로그 저장 [파일 입출력]

문제: 선박의 엔진 상태를 점검하고 그 결과를 inspection.txt 파일에 저장하는 프로그램을 작성하세요. 사용자로부터 점검 항목과 상태(정상/비정상)를 입력받아 기록합니다.



````python
from datetime import datetime

# inspection.txt 파일을 쓰기 모드로 연다.
with open("inspection.txt", "w", encoding="utf-8") as file:

    # 점검할 항목의 개수를 입력받는다.
    count = int(input("점검할 항목의 개수를 입력하세요: "))

    # 점검 항목 목록
    items = ["냉각수", "엔진오일", "연료필터"]

    # 입력한 개수만큼 반복
    for i in range(count):
        print(f"\n{i+1}번째 점검")

        # 점검 항목 선택
        while True:
            print("점검 항목을 선택하세요.")
            print("1. 냉각수")
            print("2. 엔진오일")
            print("3. 연료필터")

            choice = input("번호 입력: ")

            if choice == "1":
                item = items[0]
                break
            elif choice == "2":
                item = items[1]
                break
            elif choice == "3":
                item = items[2]
                break
            else:
                print("잘못 입력했습니다. 1~3 중에서 선택하세요.")

        # 상태 입력
        while True:
            status = input("상태(정상/비정상): ")

            if status == "정상" or status == "비정상":
                break
            else:
                print("잘못 입력했습니다. 정상 또는 비정상을 입력하세요.")

        # 현재 날짜와 시간
        current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # 파일에 저장
        file.write(f"점검시간: {current_time}, 점검항목: {item}, 상태: {status}\n")

print("\n점검 결과가 inspection.txt 파일에 저장되었습니다.")
````
